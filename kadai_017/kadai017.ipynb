```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}は{self.age}歳なので大人です")
    else:
      print(f"{self.name}は{self.age}歳なので大人ではありません")

humans = []

human1 = Human("谷", 28)
humans.append(human1)

human2 = Human("侍太郎", 15)
humans.append(human2)

human3 = Human("侍花子", 20)
humans.append(human3)

for human in humans:
  human.check_adult()
```
